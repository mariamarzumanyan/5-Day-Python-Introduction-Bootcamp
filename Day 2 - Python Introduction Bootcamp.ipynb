### Recap of Python basics related to data analysis

- Variables and Data Types: In Python, you can assign values to variables to store data. Common data types used in data analysis include integers, floats, strings, booleans, lists, and dictionaries.

- Operators: Python supports various operators for performing arithmetic, comparison, logical, and assignment operations. These operators are essential for manipulating and analyzing data.

- Data Structures: Python provides built-in data structures like lists, tuples, sets, and dictionaries. These data structures are useful for storing and organizing data in different formats.

- Control Flow: Python offers control flow statements like if-else conditions and loops (for and while) to control the execution of code based on certain conditions or iterate over a sequence of elements.

- Functions: Functions in Python allow you to encapsulate reusable blocks of code. You can define functions to perform specific tasks or operations on data, making your code more modular and maintainable.

- Input and Output: Python provides functions like input() to prompt the user for input and print() to display output on the console. These functions are crucial for interacting with users and displaying results.

- Libraries: Python has an extensive ecosystem of libraries for data analysis, such as Pandas, NumPy, and Matplotlib. These libraries provide specialized functions and tools for working with data, performing mathematical operations, and creating visualizations.

### Introduction to Data Analysis and Its Importance

Data analysis is the process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, drawing insights, and supporting decision-making. It involves applying various techniques, tools, and methodologies to extract valuable knowledge from raw data.

Data analysis plays a crucial role in numerous fields and industries, including business, finance, healthcare, marketing, social sciences, and more. Here are some reasons why data analysis is important:

- Informed Decision-Making: Data analysis enables informed decision-making by providing insights and evidence based on data-driven findings. It helps organizations and individuals make informed choices, identify trends, and evaluate potential outcomes.

- Business Intelligence: Data analysis helps businesses gain insights into customer behavior, market trends, and competitive landscapes. By analyzing data, companies can identify opportunities, optimize processes, and make data-driven decisions to drive growth and success.

- Problem Solving: Data analysis helps in identifying and solving problems by examining patterns, correlations, and anomalies within the data. It provides a systematic approach to understanding complex issues, finding root causes, and developing effective solutions.

- Identifying Trends and Patterns: Data analysis allows for the identification of trends, patterns, and relationships within large datasets. These insights can be used to uncover hidden patterns, predict future trends, and make informed predictions.

- Performance Evaluation: Data analysis enables the evaluation of performance, whether it's individual performance, business performance, or the effectiveness of strategies and initiatives. By analyzing relevant metrics and key performance indicators (KPIs), areas for improvement can be identified and appropriate actions can be taken.

- Risk Assessment and Mitigation: Data analysis helps in assessing risks and identifying potential issues before they escalate. By analyzing historical data and trends, patterns of risk can be detected, allowing for proactive measures and risk mitigation strategies.

- Data-Driven Insights: Data analysis provides valuable insights that can lead to actionable recommendations. These insights can help organizations optimize operations, improve customer experiences, and drive innovation.

- Scientific Research and Discoveries: Data analysis is essential in scientific research, enabling researchers to analyze experimental data, validate hypotheses, and draw meaningful conclusions. It aids in pushing the boundaries of knowledge and making scientific discoveries.

### Pandas

Pandas is a powerful and versatile library that provides data structures and data analysis tools for handling structured data. It is built on top of NumPy and is designed to make data manipulation and analysis more intuitive and efficient. 

- Data Structures: Pandas provides two main data structures: Series and DataFrame. Series is a one-dimensional labeled array capable of holding any data type, while DataFrame is a two-dimensional table with labeled rows and columns, similar to a spreadsheet or SQL table.

```python
# pandas has a well-known alais pd
import pandas as pd
```

We create a Series using the pd.Series() function, passing a list of values. We then print the Series, its values, and its index.

```python
# Creating a Series
s = pd.Series([3, 7, 9, 12, 5])
print("Series:")
print(s)
```

```python
# Accessing values and index of a Series
print("Values in the Series:")
print(s.values)
print("Index of the Series:")
print(s.index)
```

We create a DataFrame using the pd.DataFrame() function, passing a dictionary where keys represent column names and values represent data. We print the DataFrame, its columns, and its index.

```python
# Creating a DataFrame
data = {'Name': ['John', 'Emma', 'Sam', 'Lisa'],
        'Age': [25, 28, 31, 27],
        'City': ['New York', 'London', 'Paris', 'Sydney']}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)
```

```python
# Accessing columns and index of a DataFrame
print("Columns of the DataFrame:")
print(df.columns)
print("Index of the DataFrame:")
print(df.index)
```

```python
# Accessing specific column(s) of a DataFrame
print("Accessing a specific column:")
print(df['Name'])
```

```python
# Accessing specific row(s) of a DataFrame
print("Accessing a specific row:")
print(df.loc[2])
```

```python
# Accessing specific element(s) of a DataFrame
print("Accessing a specific element:")
print(df.loc[1, 'Age'])
```

### NumPy

NumPy (Numerical Python) is a fundamental library for scientific computing in Python. It provides powerful data structures and efficient numerical operations, making it essential for numerical computations and data manipulation. Key features of NumPy include:

- Ndarray: NumPy introduces the ndarray object, which is an efficient multi-dimensional array that allows for fast mathematical operations. The ndarray is the backbone of numerical operations in NumPy.

- Mathematical Functions: NumPy provides a wide range of mathematical functions that operate efficiently on arrays. These functions include basic arithmetic operations, trigonometric functions, statistical functions, linear algebra operations, and more.

- Broadcasting: NumPy supports broadcasting, which is a powerful mechanism for performing operations on arrays with different shapes. Broadcasting allows for efficient element-wise operations and reduces the need for explicit looping.

- Array Manipulation: NumPy offers functions for reshaping, concatenating, splitting, and transposing arrays. These functions allow for efficient manipulation and transformation of array data.

- Random Number Generation: NumPy provides functions for generating random numbers from various probability distributions. This is useful for simulating data, creating random samples, and performing Monte Carlo simulations.

```python
# similar to pandas, numpy also has an alias named np
import numpy as np
```

```python
# Creating a 1-D array
arr_1d = np.array([1, 2, 3, 4, 5])
print("1-D Array:")
print(arr_1d)
```

```python
# Creating a 2-D array
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2-D Array:")
print(arr_2d)
```

```python
# Accessing elements of an array
print("Accessing elements of the array:")
print("Element at index 2 in the 1-D array:", arr_1d[2])
print("Element at index (1, 2) in the 2-D array:", arr_2d[1, 2])
```

```python
# Array shape and size
print("Shape of the array:", arr_2d.shape)
print("Size of the array:", arr_2d.size)
```

```python
# Array arithmetic operations
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print("Array sum operations:")
print("arr1 + arr2:", arr1 + arr2)
```

```python
print("Array multiplication operations:")
print("arr1 * arr2:", arr1 * arr2)
```

```python
# Array functions and methods
print("Array functions and methods:")
print("Sum of array elements:", np.sum(arr_1d))
print("Mean of array elements:", np.mean(arr_1d))
print("Maximum element of array:", np.max(arr_1d))
```

### Data Cleaning: Handling Missing Values, Duplicates, and Outliers.

##### Duplicates

```python
# Check for duplicates
duplicate_rows = df.duplicated()
any_duplicates = duplicate_rows.any()
num_duplicates = duplicate_rows.sum()

print("Duplicate Rows:")
print(df[duplicate_rows])
print()
print("Any duplicates exist:", any_duplicates)
print("Number of duplicates:", num_duplicates)
```

```python
# Remove duplicates
df_unique = df.drop_duplicates()

print("DataFrame without duplicates:")
print(df_unique)
```

```python
# Remove duplicates based on 'Initial'
df_unique = df.drop_duplicates(subset=['Initial'])

print("DataFrame without duplicates:")
print(df_unique)
```

As you can see, the indexes of the data do not match. We can reset the index one of the two ways. 

```python
df_unique = df_unique.reset_index()
df_unique
```

```python
df_unique = df.drop_duplicates(subset=['Initial'])
df_unique.reset_index(inplace=True)
df_unique
```

Even better way to do it! In this example, the reset_index() method is called on the DataFrame df_unique. The drop=True parameter specifies that the old index should be dropped, and the inplace=True parameter ensures that the changes are made directly to the DataFrame df without creating a new DataFrame.


```python
df_unique = df.drop_duplicates(subset=['Initial'])
df_unique.reset_index(drop = True, inplace=True)
df_unique
```

### Filling missing values

- Use the fillna() method to replace missing values with specified values or strategies.
- Pass a scalar value to replace all missing values in the DataFrame.
- Pass a dictionary with column names as keys and replacement values as values to fill missing values column-wise.
- Use methods such as 'ffill' (forward fill) or 'bfill' (backward fill) to propagate non-missing values forward or backward.

```python
# Replace missing values with a specific value
df_filled = df.fillna(0)
df_filled
```

```python
# Replace missing values column-wise with different values
fill_values = {'Age': 30, 'Height': 170}
df_filled_columns = df.fillna(fill_values)
df_filled_columns
```

```python
# Forward fill missing values
df_ffilled = df.fillna(method='ffill')
df_ffilled
```

```python
# Backward fill missing values
df_bfilled = df.fillna(method='bfill')
df_bfilled
```

```python
# Perform linear regression for scatter plot 1
# Create a figure and axes
fig, ax = plt.subplots()

# Create a scatter plot
ax.scatter(df['Age'], df['Weight'])

# Set labels and title
ax.set_xlabel('Age')
ax.set_ylabel('Weight')
ax.set_title('Scatter Plot of Age and Height')


coefficients = np.polyfit(df['Age'], df['Weight'], 1)
regression_line = np.polyval(coefficients, df['Age'])
ax.plot(df['Age'], regression_line, color='blue')

# Show the plot
plt.show()
```

```python
# Create a histogram
plt.hist(df['Age'], bins=10, edgecolor='black')

# Customize the plot
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python

# Calculate the quantiles
quantiles = np.percentile(df['Age'], [25, 50, 75])

# Create a histogram
n, bins, patches = plt.hist(df['Age'], bins=10, edgecolor='black')

# Plot vertical lines for the quantiles
for q in quantiles:
    plt.axvline(q, color='red', linestyle='--')

# Customize the plot
plt.title('Histogram with Quantiles')
plt.xlabel('Age')
plt.ylabel('Frequency')


# Display the plot
plt.show()
```

```python
# Create a figure and subplot
fig, ax = plt.subplots()

# Create the boxplot on the subplot
ax.boxplot(df['Age'])

# Customize the subplot
ax.set_title('Box Plot')
ax.set_xlabel('Age')
ax.set_ylabel('Values')

# Display the plot
plt.show()
```

This example combines a histogram and a boxplot, showing the distribution of data through the histogram and the summary statistics through the boxplot.

```python
# Data for the histogram
data = np.random.normal(0, 1, 1000)

# Create a figure and subplots for histogram and boxplot
fig, (ax_hist, ax_box) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [3, 1]})

# Plot the histogram
ax_hist.hist(data, bins=10, edgecolor='black')
ax_hist.set_title('Histogram')

# Plot the boxplot
ax_box.boxplot(data, vert=False)
ax_box.set_title('Boxplot')

# Remove x-axis ticks and labels from the histogram
ax_hist.set_xticks([])

# Customize the layout and appearance of the subplots
plt.tight_layout()

# Display the plot
plt.show()
```
