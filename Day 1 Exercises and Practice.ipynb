## Day 1: Introduction to Python Basics

## Exercises and Practice

- Exercise 1: Print your name.

> Write a Python program that prints the your name to the console. Customize it using f-strings.

- Exercise 2: Calculate the Area of a Circle
> Write a Python program that calculates and prints the area of a circle. You can use the formula area = π * r^2 where π is a constant (you can use an approximation like 3.14) and r is the radius of the circle. Prompt the user to enter the radius and display the result.


- Exercise 3: Convert Celsius to Fahrenheit
>Write a Python program that converts Celsius to Fahrenheit. Prompt the user to enter a temperature in Celsius and display the equivalent temperature in Fahrenheit. The conversion formula is F = (C * 9/5) + 32.

- Exercise 4: Calculate the Sum of Digits
>Write a Python program that calculates the sum of the digits in a given integer. Prompt the user to enter an integer, calculate the sum of its digits, and display the result. For example, if the input is 12345, the sum of its digits would be 1 + 2 + 3 + 4 + 5 = 15.

- Exercise 5: Check if a Number is Even or Odd
>Write a Python program that checks if a given number is even or odd. Prompt the user to enter a number and display whether it is even or odd.

- Exercise 6: Swap Two Variables
>Write a Python program that swaps the values of two variables. Prompt the user to enter the values of two variables, swap their values, and display the updated values.

- Exercise 7: Calculate the Squared Sum of a List
>Write a Python program that calculates the squared sum of all the numbers in a given list. Prompt the user to enter a list of numbers (separated by spaces), calculate the squares then sum of the numbers, and display the result.

- Exercise 8: Print Fibonacci Series
>Write a Python program that prints the Fibonacci series up to a given number of terms. Prompt the user to enter the number of terms, calculate and display the Fibonacci series.

- Exercise 9: Check if a String is a Palindrome
> Write a Python program that checks if a given string is a palindrome. Prompt the user to enter a string, check if it reads the same forwards and backwards (ignoring spaces and capitalization), and display the result.

- Exercise 10: Count the Number of Words in a Sentence
>Write a Python program that counts the number of words in a given sentence. Prompt the user to enter a sentence, count the number of words, and display the result.
